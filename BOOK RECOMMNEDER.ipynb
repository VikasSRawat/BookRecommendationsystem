# Making Popular rating recommendation for the homepage

```python
ratings_with_name = ratings.merge(books,on='ISBN')
```

```python
ratings_with_name.shape
```

```python
# Now each book has multiple ratings from different users
ratings_with_name
```

```python
num_rating = ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating.rename(columns={'Book-Rating':'Number of ratings'},inplace = True)
num_rating
```

```python
# Now finding the average rating of each book
avg_rating = ratings_with_name.groupby('Book-Title').mean()['Book-Rating'].reset_index()
avg_rating.rename(columns={'Book-Rating':'Avg Rating'})
avg_rating
```

```python
popular_df = num_rating.merge(avg_rating,on="Book-Title")
popular_df
```

```python
# taking those books which have the top most ratings from the entire website(top 50)
popular_df=popular_df[popular_df['Number of ratings']>=150].sort_values('Book-Rating',ascending=False).head(50)
popular_df = popular_df.merge(books,on="Book-Title").drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M','Number of ratings','Book-Rating']]
```

# Making a Collabrative based recommendation system

```python
ratings_with_name.head()
```

```python
x = ratings_with_name.groupby('User-ID').count()['Book-Rating'] > 100
rated_users = x[x].index
```

```python
filtered_rating = ratings_with_name[ratings_with_name['User-ID'].isin(rated_users)]
```

```python
y = filtered_rating.groupby('Book-Title').count()['Book-Rating'] >= 50
famous_books = y[y].index
```

```python
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
final_ratings
```

```python
final_ratings.drop_duplicates()
```

```python
pt = final_ratings.pivot_table(index = 'Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pt.fillna(0,inplace = True)
```

```python
pt.tail()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores = cosine_similarity(pt)
```

```python
similarity_scores
```

```python
def recommend_book(book_name):
    index = np.where(pt.index == book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_scores[index])),key = lambda x : x[1],reverse = True)[1:6]
    for i in similar_items:
        print(pt.index[i[0]])
```

```python
recommend_book('The Da Vinci Code')
```
